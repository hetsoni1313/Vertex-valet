## Clean Description part(Remove HTML tags , HTML entities)

```python
import html
import re

def clean_description(text):
    if not text or pd.isna(text):
        return None

    text = str(text)

    # 1️⃣ Decode HTML entities (&amp; → &)
    text = html.unescape(text)

    # 2️⃣ Remove HTML tags
    text = re.sub(r"<[^>]+>", " ", text)

    # 3️⃣ Remove leftover HTML entities if any
    text = re.sub(r"&[a-zA-Z]+;", " ", text)

    # 4️⃣ Normalize whitespace
    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
df['book_description'] = df['book_description'].apply(clean_description)
```

### Normalization Author Name

```python
def clean_author(author):
    if not author or pd.isna(author):
        return None

    author = str(author).lower()

    # remove unwanted punctuation
    author = re.sub(r"[.,/]", " ", author)

    # remove brackets content
    author = re.sub(r"\(.*?\)", "", author)

    # remove numbers
    author = re.sub(r"\d+", "", author)

    # keep letters and spaces only
    author = re.sub(r"[^a-z ]", " ", author)

    # normalize spaces
    author = re.sub(r"\s+", " ", author)

    return author.strip()
```

```python
df["Author"] = df["Author"].apply(clean_author)
```
